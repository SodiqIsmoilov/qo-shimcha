```python
#EKUK(a,b,c)
a = int(input('a soni : '))
b = int(input('b soni : '))
c = int(input('c soni : '))
ekuk = 0
k=a
if b>=c>=a or c>=b>=a:
    k=a
elif c>=a>=b or a>=c>=b:
    k=b
elif a>=b>=c or b>=a>=c:
    k=c
for i in range(k, a*b*c+1):
    if i%a==0 and i%b==0 and i%c==0:
        ekuk=i
        break
print(ekuk)
```
